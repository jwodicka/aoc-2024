```
var lines = data.Split("\n");

List<int[]> keys = [];
List<int[]> locks = [];

for(int i = 0; i < lines.Length; i += 8) {
    bool isLock = lines[i] == "#####";
    int[] pattern = [0, 0, 0, 0, 0];
    for(int j = 1; j < 6; j++) {
        string line = lines[i + j];
        for(int k = 0; k < 5; k++) {
            pattern[k] += line[k] == '#' ? 1 : 0;
        }
    }
    if (isLock) {
        locks.Add(pattern);
    } else {
        keys.Add(pattern);
    }
}

display(locks);
display(keys);
```

```
bool MightFit(int[] lockPattern, int[] keyPattern) {
    for (int i = 0; i < 5; i++) {
        if (lockPattern[i] + keyPattern[i] > 5) {
            return false;
        }
    }
    return true;
}

int fits = 0;
foreach(var lockPattern in locks) {
    foreach(var keyPattern in keys) {
        if (MightFit(lockPattern, keyPattern)) { fits++; }
    }
}

fits
```
