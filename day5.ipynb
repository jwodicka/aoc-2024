### Parse out rules and update lists

```
Regex ruleRegex = new(@"^(\d+)\|(\d+)$", RegexOptions.Multiline);

List<(int Pre, int Post)> rules = [];

foreach (Match match in ruleRegex.Matches(data)) {
    int pre = int.Parse(match.Groups[1].Value);
    int post = int.Parse(match.Groups[2].Value);
    rules.Add((pre, post));
}

Regex updateRegex = new(@"^(?:\d+,)*\d+$", RegexOptions.Multiline);

List<List<int>> updates = [];
foreach (Match match in updateRegex.Matches(data)) {
    updates.Add(match.Value.Split(",").Select(s => int.Parse(s)).ToList());
}

foreach(var l in updates) {
    Print(l.DebugPrint());
}
```

```
bool IsUpdateValid(List<int> update) {
    foreach(var (pre, post) in rules) {
        if (update.Contains(pre) && update.Contains(post)) {
            // This rule applies!
            if (update.IndexOf(pre) > update.IndexOf(post)) {
                return false;
            }
        }
    }
    return true;
}

var validUpdates = updates.Where(IsUpdateValid);
```

```
int MiddleElement(List<int> update) {
    return update[update.Count / 2];
}

var middleElements = validUpdates.Select(MiddleElement);

Print(middleElements.DebugPrint());

Print(middleElements.Sum());
```

```
var invalidUpdates = updates.Where(v => !IsUpdateValid(v));

List<int> MakeUpdateMoreValid(List<int> update) {
    foreach(var (pre, post) in rules) {
        if (update.Contains(pre) && update.Contains(post)) {
            // This rule applies! Make an improvement
            int preIndex = update.IndexOf(pre);
            int postIndex = update.IndexOf(post);
            // [a a a post b b b pre c c c]
            // [a a a pre post b b b c c c]
            if (preIndex > postIndex) {
                // Print((pre, preIndex, post, postIndex));
                return [
                    .. update.Take(postIndex),
                    pre,
                    .. update.Skip(postIndex + 1).Take(preIndex - postIndex - 1),
                    post,
                    .. update.Skip(preIndex + 1)
                ];
            }
        }
    }
    return update;
}

List<int> MakeUpdateValid(List<int> initialUpdate) {
    List<int> update = initialUpdate;
    while(!IsUpdateValid(update)) {
        update = MakeUpdateMoreValid(update);
    }
    return update;
}

var validatedUpdates = invalidUpdates.Select(MakeUpdateValid);

// foreach (var update in validatedUpdates) {
//     Print(update.DebugPrint());
// }

var middleElements = validatedUpdates.Select(MiddleElement);

// Print(middleElements.DebugPrint());

Print(middleElements.Sum());

```
