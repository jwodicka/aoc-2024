# Part 1

```
static Dictionary<string, HashSet<string>> Connections = [];
var pairs = data.Split('\n').Select(s => s.Split('-'));

void Connect(string a, string b) {
    if (!Connections.TryGetValue(a, out var connected)) {
        Connections[a] = connected = [];
    }
    connected.Add(b);
}

foreach (var pair in pairs) {
    Connect(pair[0], pair[1]);
    Connect(pair[1], pair[0]);
}
```

```
var machines = Connections.Keys.Order().ToList();
```

```
bool AreTriad(string a, string b, string c) =>
    Connections[a].Contains(b) &&
    Connections[a].Contains(c) &&
    Connections[b].Contains(c);

HashSet<(string, string, string)> triads = [];
for (int i = 0; i < machines.Count; i++) {
    for (int j = i+1; j < machines.Count; j++) {
        for (int k = j+1; k < machines.Count; k++) {
            var (a, b, c) = (machines[i], machines[j], machines[k]);
            if (AreTriad(a, b, c)) {
                triads.Add((a, b, c));
            }
        }
    }
}

bool HasT((string a, string b, string c) triad) =>
    triad.a.StartsWith("t") ||
    triad.b.StartsWith("t") ||
    triad.c.StartsWith("t");

triads.Where(HasT).Count()
```

# Part 2

```
// One machine is a mutual group of itself.
// To add another machine, check each group; if the added machine is connected to all machines in the group, it joins.
// Also add itself as a singleton group.

List<HashSet<string>> mutualGroups = [];

public static bool IsFullyConnected(this string self, IEnumerable<string> others) {
    foreach(string other in others) {
        if(!Connections[self].Contains(other)) {
            return false;
        }
    }
    return true;
}

foreach (string machine in machines) {
    List<HashSet<string>> newGroups = [[machine]];
    foreach (var group in mutualGroups) {
        if (machine.IsFullyConnected(group)) {
            newGroups.Add([machine, ..group]);
        }
    }
    mutualGroups.AddRange(newGroups);
}

var biggest = mutualGroups.MaxBy(group => group.Count);
string.Join(",", biggest.Order())
```
