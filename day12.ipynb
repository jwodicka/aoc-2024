# Part 1

```
string[] stringMap = data.Split("\n");
char[][] map = stringMap.Select(
    line => line.ToCharArray()
).ToArray();

int Height = map.Length;
int Width = map[0].Length;
Coord Size = (Width, Height);
```

```
class Region {
    public char[][] Map { get; init; }
    public HashSet<Coord> Cells { get; init; }
    public required char Value { get; init; }

    static Coord[] CardinalDirections = [(0, +1), (0, -1), (+1, 0), (-1, 0)];

    public int Area => Cells.Count;
    public int Perimeter {
        get {
            int result = 0;
            foreach (Coord cell in Cells) {
                HashSet<Coord> neighbors = 
                    [..CardinalDirections.Select(c => Add(c, cell))];
                result += neighbors.Except(Cells).Count();
            }
            return result;
        }
    }
    public int Price => Area * Perimeter;
}

HashSet<Coord> ToVisit = [];
for(int r = 0; r < Height; r++) {
    for (int c = 0; c < Width; c++) {
        ToVisit.Add((r, c));
    }
}
```

```
List<Region> regions = [];

while (ToVisit.Count > 0) {
    Coord seed = ToVisit.First();
    ToVisit.Remove(seed);
    char value = map[seed.R][seed.C];
    Region region = new Region() {
        Map = map,
        Cells = [seed],
        Value = value,
    };

    void ProcessCell(Coord cell) {
        var neighbors = CardinalDirections.Select(c => Add(c, cell))
            .Where(c => IsInBounds(c, Size))
            .Where(ToVisit.Contains)
            .Where(c => map[c.R][c.C] == value);
        foreach (Coord neighbor in neighbors) {
            ToVisit.Remove(neighbor);
            region.Cells.Add(neighbor);
            ProcessCell(neighbor);
        }
    }
    ProcessCell(seed);
    regions.Add(region);
}

int total = 0;
foreach (var region in regions) {
    total += region.Price;
}
total
```

# Part 2

```
using Fence = (Direction D, Coord C);

enum Direction {
    Right = 0,
    Left = 1,
    Down = 2,
    Up = 3,
}

class Region {
    public char[][] Map { get; init; }
    public HashSet<Coord> Cells { get; init; }
    public required char Value { get; init; }

    static Coord[] CardinalDirections = [(0, +1), (0, -1), (+1, 0), (-1, 0)];

    public int Area => Cells.Count;
    public int Perimeter {
        get {
            IEnumerable<Fence> FenceDirections = Enum.GetValues<Direction>().Zip(CardinalDirections);
            
            HashSet<Fence> allFences = [];
            foreach (Coord cell in Cells) {
                HashSet<Fence> fences = 
                    [..FenceDirections.Select(f => (f.D, (cell)))];
        
                fences.RemoveWhere(f => Cells.Contains(Add(f.C, CardinalDirections[(int)f.D])));
                allFences.UnionWith(fences);
            }
            // display(allFences);
            // Deduplicate parallel fences

            int height = Map.Length;
            int width = Map[0].Length;

            List<Direction> Horizontal = [Direction.Up, Direction.Down];
            List<Direction> Vertical = [Direction.Left, Direction.Right];

            
            for (int r = 0; r < height; r++) {
                for (int c = 0; c < width; c++) {
                    foreach (Direction d in Horizontal) {
                        Fence fence = (d, (r, c));
                        Fence neighbor = (d, (r, c+1));
                        if (allFences.Contains(neighbor)) {
                            allFences.Remove(fence);
                        }
                    }
                    foreach (Direction d in Vertical) {
                        Fence fence = (d, (r, c));
                        Fence neighbor = (d, (r+1, c));
                        if (allFences.Contains(neighbor)) {
                            allFences.Remove(fence);
                        }
                    }
                }
            }

            return allFences.Count;
        }
    }
    public int Price => Area * Perimeter;
}

HashSet<Coord> ToVisit = [];
for(int r = 0; r < Height; r++) {
    for (int c = 0; c < Width; c++) {
        ToVisit.Add((r, c));
    }
}
```

```
List<Region> regions = [];

while (ToVisit.Count > 0) {
    Coord seed = ToVisit.First();
    ToVisit.Remove(seed);
    char value = map[seed.R][seed.C];
    Region region = new Region() {
        Map = map,
        Cells = [seed],
        Value = value,
    };

    void ProcessCell(Coord cell) {
        var neighbors = CardinalDirections.Select(c => Add(c, cell))
            .Where(c => IsInBounds(c, Size))
            .Where(ToVisit.Contains)
            .Where(c => map[c.R][c.C] == value);
        foreach (Coord neighbor in neighbors) {
            ToVisit.Remove(neighbor);
            region.Cells.Add(neighbor);
            ProcessCell(neighbor);
        }
    }
    ProcessCell(seed);
    regions.Add(region);
}

int total = 0;
foreach (var region in regions) {
    total += region.Price;
}
total
```
