```
public static BigInteger Sum(this IEnumerable<BigInteger> self) {
    BigInteger acc = 0;
    foreach (var i in self) {
        acc += i;
    }
    return acc;
}
```

```
BigInteger CountCombos(string pattern) {
    if (pattern == "") {
        return 1;
    }
    BigInteger result = 0;
    foreach (string towel in towels) {
        if (pattern.StartsWith(towel)) {
            result += MemoizedCountCombos(pattern.Substring(towel.Length));
        }
    }
    return result;
}

Dictionary<string, BigInteger> Combos = [];
BigInteger MemoizedCountCombos(string pattern) {
    if (!Combos.ContainsKey(pattern)) {
        Combos.Add(pattern, CountCombos(pattern));
    }
    return Combos[pattern];
}

MemoizedCountCombos(patterns[0]);

patterns.Select(p => MemoizedCountCombos(p)).Sum()
```
