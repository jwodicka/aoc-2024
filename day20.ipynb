```
Coord FindSymbol(char[][] map, char symbol) {
    for(int r = 0; r < map.Length; r++) {
        for(int c = 0; c < map[r].Length; c++) {
            if (map[r][c] == symbol) {
                return (r, c);
            }
        }
    }
    throw new System.InvalidOperationException("No Symbol?!");
}
```

```
using Path = (System.Collections.Generic.List<Coord> route, int cost);
```

```
(Path, Dictionary<Coord, List<Path>>) BestPossiblePath(Coord start, Coord end, char[][] map) {
    Dictionary<Coord, List<Path>> BestPathsSeen = [];
    PriorityQueue<Coord, int> PositionsToEvaluate = new();

    char MapAt(Coord c) => map[c.R][c.C];
    
    BestPathsSeen.Add(end, [([end], 0)]);
    PositionsToEvaluate.Enqueue(end, 0);

    List<(Coord, int)> FromPositions(Coord current) {
        List<(Coord, int)> result = [];

        foreach(Direction d in Enum.GetValues<Direction>()) {
            Coord pos = Add(current, Vector(d));
            if (MapAt(pos) != '#') {
                result.Add((pos, 1));
            } 
        }
        return result;
    } 

    Stopwatch timer = Stopwatch.StartNew();
    while(PositionsToEvaluate.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(60)) {
        Coord p = PositionsToEvaluate.Dequeue();

        List<Path> currentPaths = BestPathsSeen[p];
        int currentValue = currentPaths.First().cost;
        if (p == start) {
            return (currentPaths.Single(), BestPathsSeen);
        }

        foreach(var (position, cost) in FromPositions(p)) {
            List<Path> newPaths = currentPaths.Select<Path, Path>(
                cp => ([position, ..cp.route], cp.cost + cost)
            ).ToList();

            if (!BestPathsSeen.ContainsKey(position)) {
                BestPathsSeen.Add(position, newPaths);
                PositionsToEvaluate.Enqueue(position, currentValue + cost);
            } else {
                List<Path> knownPaths = BestPathsSeen[position];
                int knownValue = knownPaths.First().cost;
                if (knownValue > currentValue + cost) {
                    // Replace existing paths.
                    BestPathsSeen[position] = newPaths;
                    PositionsToEvaluate.Enqueue(position, currentValue + cost);
                } else if (knownValue == currentValue + cost) {
                    // Append to existing paths.
                    BestPathsSeen[position] = knownPaths.Concat(newPaths).ToList();
                }
            }
        }
    }

    throw new ArgumentException();
}

// BestPossiblePath(FindSymbol(initialMap, 'S'), FindSymbol(initialMap, 'E'), initialMap)
```

```
int FindCheats(Path path, char[][] map, Dictionary<Coord, List<Path>> bestPathSeen, int minCheat = 100) {
    Coord size = (map.Length, map[0].Length);
    display(size);
    char MapAt(Coord c) => map[c.R][c.C];

    int CostAt(Coord c) => bestPathSeen[c].First().cost;

    int result = 0;
    foreach(Coord point in path.route) {
        var cost = CostAt(point);
        
        // Get list of cheatable cells:
        //   * 
        //  *N*
        // *W@E*
        //  *S*
        //   *
        //
        //   #.
        //   .#
        List<Coord> cheatTargetOffsets = [(-2, 0), (-1, +1), (0, +2), (+1, +1), (+2, 0), (+1, -1), (0, -2), (-1, -1)];
        var cheatableCells = cheatTargetOffsets.Select(o => Add(o, point)).Where(p => IsInBounds(p, size));
        
        // Print(cheatableCells.DebugPrint());
        cheatableCells = cheatableCells.Where(p => MapAt(p) != '#');

        // Filter by cells that are better than our natural path.
        var goodCheats = cheatableCells.Where(p => CostAt(p) < cost - minCheat);
        foreach (var cheat in goodCheats) {
            result++;
        }
    }
    return result;
}
```

```
int FindBigCheats(Path path, char[][] map, Dictionary<Coord, List<Path>> bestPathSeen, int minCheat = 100) {
    Coord size = (map.Length, map[0].Length);
    display(size);
    char MapAt(Coord c) => map[c.R][c.C];

    int CostAt(Coord c) => bestPathSeen[c].First().cost;

    IEnumerable<(Coord p, int cost)> GetCheatableCells(Coord start) {
        List<(Coord p, int cost)> offsets = [];
        for(int r = -20; r <= 20; r++) {
            int cSpan = 20 - Math.Abs(r);
            for(int c = -cSpan; c <= cSpan; c++) {
                offsets.Add(((r, c), Math.Abs(r)+Math.Abs(c)));
            }
        }

        var cheatableCells = offsets.Select(o => (p: Add(o.p, start), o.cost)).Where(cell => IsInBounds(cell.p, size));
        
        // Print(cheatableCells.DebugPrint());
        cheatableCells = cheatableCells.Where(cell => MapAt(cell.p) != '#');
        return cheatableCells;
    }

    int result = 0;
    foreach(Coord point in path.route) {
        var cost = CostAt(point);
        
        // Filter by cells that are better than our natural path.
        var goodCheats = GetCheatableCells(point).Where(cell => CostAt(cell.p)+cell.cost-1 < cost - minCheat);
        foreach (var cheat in goodCheats) {
            result++;
        }
    }
    return result;
}
```
