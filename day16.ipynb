```
#!import coord.ipynb

enum Direction {
    Up = 0,
    Right = 1,
    Down = 2,
    Left = 3
}
Coord[] Step = [(-1, 0), (0, +1), (+1, 0), (0, -1)];
Coord Vector(Direction d) => Step[(int)d];
Direction RotateCW(Direction d) => (Direction) (((int)d + 1) % 4);
Direction RotateCCW(Direction d) => (Direction) (((int)d + 4 - 1) % 4); 
```

```
Coord FindSymbol(char[][] map, char symbol) {
    for(int r = 0; r < map.Length; r++) {
        for(int c = 0; c < map[r].Length; c++) {
            if (map[r][c] == symbol) {
                return (r, c);
            }
        }
    }
    throw new System.InvalidOperationException("No Symbol?!");
}
```

```
using Position = (Coord P, Direction D);
```

```
int BestPossibleScore(Position start, Coord end, char[][] map) {
    Dictionary<Position, int> BestValueSeen = [];
    PriorityQueue<Position, int> PositionsToEvaluate = new();

    char MapAt(Coord c) => map[c.R][c.C];

    List<Position> EndPositions = [
        (end, Direction.Up), (end, Direction.Down), (end, Direction.Left), (end, Direction.Right)
    ];
    
    foreach (var endP in EndPositions) {
        BestValueSeen.Add(endP, 0);
        PositionsToEvaluate.Enqueue(endP, 0);
    }

    List<(Position, int)> FromPositions(Position current) {
        List<(Position, int)> result = [];
        
        if (MapAt(Subtract(current.P, Vector(current.D))) != '#') {
            result.Add(((Subtract(current.P, Vector(current.D)), current.D), 1));
        }
        result.Add(((current.P, RotateCW(current.D)), 1000));
        result.Add(((current.P, RotateCCW(current.D)), 1000));
        return result;
    } 

    Stopwatch timer = Stopwatch.StartNew();
    while(PositionsToEvaluate.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(60)) {
        Position p = PositionsToEvaluate.Dequeue();

        int currentValue = BestValueSeen[p];
        if (p == start) {
            return currentValue;
        }

        foreach(var (position, cost) in FromPositions(p)) {
            if (!BestValueSeen.ContainsKey(position)) {
                BestValueSeen.Add(position, currentValue + cost);
                PositionsToEvaluate.Enqueue(position, currentValue + cost);
            } else if (BestValueSeen[position] > currentValue + cost) {
                BestValueSeen[position] = currentValue + cost;
                PositionsToEvaluate.Enqueue(position, currentValue + cost);
            }
        }
    }
    return -1;
}
```

```
// REAL VERSION

HashSet<Coord> BestPathCells(Position start, Coord end, char[][] map) {
    Dictionary<Position, List<Path>> BestPathsSeen = [];

    char MapAt(Coord c) => map[c.R][c.C];

    List<Position> EndPositions = [
        (end, Direction.Up), (end, Direction.Down), (end, Direction.Left), (end, Direction.Right)
    ];
    
    List<(Position, int)> FromPositions(Position current) {
        List<(Position, int)> result = [];
        
        if (MapAt(Subtract(current.P, Vector(current.D))) != '#') {
            result.Add(((Subtract(current.P, Vector(current.D)), current.D), 1));
        }
        result.Add(((current.P, RotateCW(current.D)), 1000));
        result.Add(((current.P, RotateCCW(current.D)), 1000));
        return result;
    } 

    List<Path> ComputeBestPath() {
        BestPathsSeen.Clear();
        PriorityQueue<Position, int> PositionsToEvaluate = new();
        foreach (var endP in EndPositions) {
            BestPathsSeen.Add(endP, [([endP], 0)]);
            PositionsToEvaluate.Enqueue(endP, 0);
        }

        List<Path> result = [];

        Stopwatch timer = Stopwatch.StartNew();
        while(PositionsToEvaluate.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(60)) {
            Position p = PositionsToEvaluate.Dequeue();

            List<Path> currentPaths = BestPathsSeen[p];
            int currentValue = currentPaths.First().cost;
            if (p == start) {
                Print("Evaluating start");
                int resultValue = result.Count > 0 ? result.First().cost : currentValue + 100;
                if (currentValue < resultValue) {
                    result = currentPaths;
                } else if (currentValue == resultValue) {
                    result.AddRange(currentPaths);
                }
            }

            foreach(var (position, cost) in FromPositions(p)) {
                List<Path> newPaths = currentPaths.Select<Path, Path>(
                    cp => ([position, ..cp.route], cp.cost + cost)
                ).ToList();

                if (!BestPathsSeen.ContainsKey(position)) {
                    BestPathsSeen.Add(position, newPaths);
                    PositionsToEvaluate.Enqueue(position, currentValue + cost);
                } else {
                    List<Path> knownPaths = BestPathsSeen[position];
                    int knownValue = knownPaths.First().cost;
                    if (knownValue > currentValue + cost) {
                        // Replace existing paths.
                        BestPathsSeen[position] = newPaths;
                        PositionsToEvaluate.Enqueue(position, currentValue + cost);
                    } else if (knownValue == currentValue + cost) {
                        // Append to existing paths.
                        BestPathsSeen[position] = knownPaths.Concat(newPaths).ToList();
                    }
                }
            }
        }

        return result;
    }
    
    List<Path> bestPaths = ComputeBestPath();
    display(bestPaths);
    HashSet<Coord> bestTiles = [];
    foreach(Path path in bestPaths) {
        bestTiles.UnionWith(path.route.Select(pos => pos.P));
    }
    return bestTiles;
}
```
