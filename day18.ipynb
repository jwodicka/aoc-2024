```
int BestPossibleScore(Coord start, Coord end, char[][] map) {
    Dictionary<Coord, int> BestValueSeen = [];
    PriorityQueue<Coord, int> PositionsToEvaluate = new();

    char MapAt(Coord c) => map[c.Y][c.X];


    List<(Coord, int)> FromPositions(Coord current) {
        List<(Coord, int)> result = [];

        foreach(Direction d in Enum.GetValues(typeof(Direction))) {
            Coord p = Add(current, Vector(d));
            if (IsInBounds(p, Size) && MapAt(p) != '#') {
                result.Add((p, 1));
            }
        }
        return result;
    }

    BestValueSeen.Add(end, 0);
    PositionsToEvaluate.Enqueue(end, 0);

    Stopwatch timer = Stopwatch.StartNew();
    while(PositionsToEvaluate.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(60)) {
        Coord p = PositionsToEvaluate.Dequeue();

        int currentValue = BestValueSeen[p];
        if (p == start) {
            return currentValue;
        }

        foreach(var (position, cost) in FromPositions(p)) {
            if (!BestValueSeen.ContainsKey(position)) {
                BestValueSeen.Add(position, currentValue + cost);
                PositionsToEvaluate.Enqueue(position, currentValue + cost);
            } else if (BestValueSeen[position] > currentValue + cost) {
                BestValueSeen[position] = currentValue + cost;
                PositionsToEvaluate.Enqueue(position, currentValue + cost);
            }
        }
    }
    return -1;
}

BestPossibleScore((0,0), Subtract(Size, (1,1)), map)
```
