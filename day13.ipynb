```
int? NaiveMachineCost(Machine machine) => NaiveMachineCost(machine, (0, 0));

int? NaiveMachineCost(Machine machine, Coord current) {
    Print($"Checking {current}");
    if (current.R > machine.Target.R || current.C > machine.Target.C) {
        // Overshoot
        return null;
    }
    if (current == machine.Target) {
        // No additional cost; we arrived.
        return 0;
    }
    int? aCost = NaiveMachineCost(machine, Add(current, machine.A));
    int? bCost = NaiveMachineCost(machine, Add(current, machine.B));

    if (aCost is null && bCost is null) {
        return null;
    }
    if (aCost is not null) {
        aCost += 3;
    }
    if (bCost is not null) {
        bCost += 1;
    }
    if (aCost is null || bCost is null) {
        return aCost ?? bCost;
    }
    return Math.Min((int)aCost, (int)bCost);
}

```

```
using System.Numerics;
using State = (Coord Coord, int Cost);

double Length(Coord coord) => Math.Sqrt((double)(coord.C * coord.C * coord.R * coord.R));
Vector2 AsV2(Coord c) => new Vector2(c.C, c.R);

State startState = ((0,0), 0);

int? MachineCost(Machine machine) {
    HashSet<State> seenStates = [];
    PriorityQueue<State, int> queue = new();

    float aLen = AsV2(machine.A).Length();
    float bLen = AsV2(machine.B).Length();
    float aPower = aLen / 3f;
    float bPower = bLen;
    float minPower = Math.Min(aPower, bPower);
    float maxPower = Math.Max(aPower, bPower);
    int StateQuality(State state) {
        float distanceToTarget = Vector2.Distance(AsV2(state.Coord), AsV2(machine.Target));
        int distanceScore = (int)Math.Ceiling(distanceToTarget / maxPower);
        return distanceScore + state.Cost;
    }

    void Enqueue(State state) {
        if (seenStates.Contains(state)) {
            return;
        }
        seenStates.Add(state);
        queue.Enqueue(state, StateQuality(state));
    }

    Enqueue(startState);

    Stopwatch timer = Stopwatch.StartNew();
    while(queue.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(30)) {
        State state = queue.Dequeue();
        // Print($"Evaluating {state} Q:{queue.Count}");
        if (state.Coord == machine.Target) {
            return state.Cost;
        }
        foreach(State s in EvaluateState(machine, state)) {
            Enqueue(s);
        }
    }
    return null;
}

IEnumerable<State> EvaluateState(Machine machine, State state) {
    List<State> result = [ 
        (Add(state.Coord, machine.A), state.Cost + 3),
        (Add(state.Coord, machine.B), state.Cost + 1)
    ];
    // Remove overshoots
    return result.Where(s => {
        Coord current = s.Coord;
        return current.R <= machine.Target.R && current.C <= machine.Target.C;
    });
}
```

```
using State128 = (Coord128 Coord, int Cost);
using Machine128 = (Coord128 A, Coord128 B, Coord128 Target);
using System.Numerics;

public static Coord128 Add(Coord a, Coord128 b) => (a.R + b.R, a.C + b.C);
public static Coord128 Add(Coord128 a, Coord128 b) => (a.R + b.R, a.C + b.C);
public static Coord128 Scale(Coord c) => (c.R, c.C);

Coord128 offset = (10000000000000, 10000000000000);
List<Machine128> bigMachines = machines.Select(
    m => (Scale(m.A), Scale(m.B), (Add(m.Target, offset)))
).ToList();

double Length(Coord128 coord) => Math.Sqrt((double)(coord.C * coord.C * coord.R * coord.R));
double Distance(Coord128 a, Coord128 b) {
    Coord128 distance = (Int128.Abs(a.R - b.R), Int128.Abs(a.C - b.C));
    return Length(distance);
}

State128 startState = ((0,0), 0);

int? BigMachineCost(Machine128 machine) {
    HashSet<State128> seenStates = [];
    PriorityQueue<State128, int> queue = new();

    double aLen = Length(machine.A);
    double bLen = Length(machine.B);
    double aPower = aLen / 3.0;
    double bPower = bLen;
    double minPower = Math.Min(aPower, bPower);
    double maxPower = Math.Max(aPower, bPower);
    int StateQuality(State128 state) {
        double distanceToTarget = Distance(state.Coord, machine.Target);
        int distanceScore = (int)Math.Ceiling(distanceToTarget / maxPower);
        return distanceScore + state.Cost;
    }

    void Enqueue(State128 state) {
        if (seenStates.Contains(state)) {
            return;
        }
        seenStates.Add(state);
        queue.Enqueue(state, StateQuality(state));
    }

    Enqueue(startState);

    Stopwatch timer = Stopwatch.StartNew();
    while(queue.Count > 0 && timer.Elapsed < TimeSpan.FromSeconds(30)) {
        State128 state = queue.Dequeue();
        // Print($"Evaluating {state} Q:{queue.Count}");
        if (state.Coord == machine.Target) {
            return state.Cost;
        }
        foreach(State128 s in EvaluateState(machine, state)) {
            Enqueue(s);
        }
    }
    return null;
}

IEnumerable<State128> EvaluateState(Machine128 machine, State128 state) {
    List<State128> result = [ 
        (Add(state.Coord, machine.A), state.Cost + 3),
        (Add(state.Coord, machine.B), state.Cost + 1)
    ];
    // Remove overshoots
    return result.Where(s => {
        Coord128 current = s.Coord;
        return current.R <= machine.Target.R && current.C <= machine.Target.C;
    });
}
```

```
public static Int128 Sum(this IEnumerable<Int128> self) {
    Int128 acc = 0;
    foreach (var i in self) {
        acc += i;
    }
    return acc;
}

Int128? AnalyticSolve(Machine128 machine) {
    var aPress = machine.A;
    var bPress = machine.B;
    var target = machine.Target;

    // Via application of Cramer's rule (which I can use but not really understand):

    var A1 = aPress.C;
    var A2 = aPress.R;
    var B1 = bPress.C;
    var B2 = bPress.R;
    var C1 = target.C;
    var C2 = target.R;

    var aPresses = Int128.DivRem((C1 * B2) - (B1 * C2), (A1 * B2) - (B1 * A2));
    var bPresses = Int128.DivRem((A1 * C2) - (C1 * A2), (A1 * B2) - (B1 * A2));
    // display((aPresses, bPresses));
    if (aPresses.Remainder != 0 || bPresses.Remainder != 0) {
        return null;
    }
    return (aPresses.Quotient * 3) + (bPresses.Quotient);
}
var rawResult = bigMachines.Select(AnalyticSolve);
Print(rawResult.Count());
var result = rawResult.Where(i => i is not null);
Print(result.Count());
result.Cast<Int128>().Sum()
```
