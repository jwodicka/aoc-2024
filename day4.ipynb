```
using System.Linq;

#!set --name rawdata --value @value:data
var data = rawdata.ReplaceLineEndings("\n");


string[] lines = data.Split("\n", StringSplitOptions.RemoveEmptyEntries & StringSplitOptions.TrimEntries);

// Trim empty lines even though Split was supposed to do that.
lines = lines.Where(l => l.Length > 0).ToArray();

Print(lines[0].Length);
Print(lines.DebugPrint());
```

```
int width = lines[0].Length;
int height = lines.Length;

string target = "XMAS";

int CheckPath(int r, int c, int dr, int dc) {
    StringBuilder path = new StringBuilder();
    for (int d = 0; d < target.Length; d++) {
        int rCalc = r + (dr*d);
        int cCalc = c + (dc*d);
        if (rCalc >= height || cCalc >= width) {
            return 0;
        }
        if (rCalc < 0 || cCalc < 0) {
            return 0;
        }
        path.Append(lines[rCalc][cCalc]);
    }
    if (path.ToString() == target) return 1;
    return 0;
}

int FindStartingAt(int r, int c) {
    int result = 0;
    if (lines[r][c] != target[0]) {
        return result;
    }
    result += CheckPath(r, c, -1,  0); // Up
    result += CheckPath(r, c, -1, +1); // Up-Right
    result += CheckPath(r, c,  0, +1); // Right
    result += CheckPath(r, c, +1, +1); // Down-Right
    result += CheckPath(r, c, +1,  0); // Down
    result += CheckPath(r, c, +1, -1); // Down-Left
    result += CheckPath(r, c,  0, -1); // Left
    result += CheckPath(r, c, -1, -1); // Up-Left

    return result;
}

int found = 0;
for (int r = 0; r < height; r++) {
    for (int c = 0; c < width; c++) {
        found += FindStartingAt(r, c);
    }
}
Print(found);
```

```
bool isValid (char A, char B) => (A, B) switch {
    ('M', 'S') => true,
    ('S', 'M') => true,
            _  => false
};

int FindMasStartingAt(int r, int c) {
    if (lines[r][c] != 'A') return 0;

    char UL = lines[r-1][c-1];
    char UR = lines[r-1][c+1];
    char DL = lines[r+1][c-1];
    char DR = lines[r+1][c+1];

    if (isValid(UL, DR) && isValid(DL, UR)) return 1;

    return 0; 
}

int found2 = 0;
for (int r = 1; r < height -1; r++) {
    for (int c = 1; c < width -1; c++) {
        found2 += FindMasStartingAt(r, c);
    }
}
Print(found2);
```
