```
public int[] Compact(int[] disk) {
    int[] result = [.. disk];

    void Swap(int a, int b) => (result[a], result[b]) = (result[b], result[a]);


    int rPtr = result.Length - 1;
    void AdvanceRPtr() {
        while(result[rPtr] < 0 && rPtr > 0) { rPtr--; }
    }

    AdvanceRPtr();
    for(int fPtr = 0; fPtr < rPtr; fPtr++) {
        if (result[fPtr] < 0) {
            Swap(fPtr, rPtr);
            AdvanceRPtr();
        }
    }
    return result;
}

var cd = Compact(disk);
```

```
public Int128 Checksum(int[] disk) {
    Int128 result = 0;
    for (int i = 0; i < disk.Length; i++) {
        if (disk[i] >= 0) {
            result += (Int128)i * (Int128)disk[i];
        }
    }
    return result;
}

Print(Checksum(cd));
```

# Part 2

```
using Block = (int Length, int? FileId);

public List<Block> BuildDiskIndex() {
    int fileId = 0;
    List<Block> diskIndex = [];

    for(int i = 0; i < parsed.Length; i += 2) {
        int fLen = parsed[i];
        diskIndex.Add((fLen, fileId++));

        if (i+1 < parsed.Length) {
            int skipLen = parsed[i+1];
            if (skipLen > 0) {
                diskIndex.Add((skipLen, null));
            }
        }
    }
    return diskIndex;
}

List<Block> index = BuildDiskIndex();
```

```
public List<Block> Compact(List<Block> index) {
    List<Block> result = [.. index];

    void Swap(int a, int b) => (result[a], result[b]) = (result[b], result[a]);

    for(int rPtr = result.Count - 1; rPtr > 0; rPtr--) {
        Block currBlock = result[rPtr];
        if (currBlock.FileId is null) {
            continue;
        }
        void FindGap() {
            for(int fPtr = 0; fPtr < rPtr; fPtr++) {
                Block swapBlock = result[fPtr];
                if (swapBlock.FileId is null && swapBlock.Length >= currBlock.Length) {
                    int fragLength = swapBlock.Length - currBlock.Length;
                    if (fragLength > 0) {
                        result[fPtr] = (currBlock.Length, null);
                        result.Insert(fPtr + 1, (fragLength, null));
                        rPtr++;
                    }
                    // Print($"Swapping {rPtr}:{result[rPtr]}, {fPtr}:{result[fPtr]}");
                    Swap(fPtr, rPtr);
                    return;
                }
            }
        }
        FindGap();
    }
    return result;
}

var ci = Compact(index);
Print(ci.DebugPrint());
```

```
int[] Unpack(List<Block> index) {
    int diskSize = index.Select(b => b.Length).Sum();
    int[] disk = new int[diskSize];
    Array.Fill(disk, -999);

    int dPtr = 0;
    foreach(var (length, bId) in index) {
        int id = bId ?? -1;
        for (int i = 0; i < length; i++, dPtr++) {
            disk[dPtr] = id;
        }
    }
    return disk;
}

Print(Checksum(Unpack(ci)));
```
