# Part 1

```
Regex valueRegex = new(@"^(\w{3}): ([0|1])$", RegexOptions.Multiline);
Regex gateRegex = new(@"^(\w{3}) (AND|OR|XOR) (\w{3}) -> (\w{3})$", RegexOptions.Multiline);

Dictionary<string, bool> InitialValues = [];
foreach (Match match in valueRegex.Matches(data)) {
    string gate = match.Groups[1].Value;
    string value = match.Groups[2].Value;

    InitialValues[gate] = value == "0" ? false : true;
}
```

```
enum Operation {
    AND,
    OR,
    XOR
}
```

```
using Gate = (Operation op, string a, string b);
```

```
Dictionary<string, Gate> InitialGates = [];
foreach (Match match in gateRegex.Matches(data)) {
    string a = match.Groups[1].Value;
    Operation operation = Enum.Parse<Operation>(match.Groups[2].Value);
    string b = match.Groups[3].Value;
    string gate = match.Groups[4].Value;

    InitialGates[gate] = (operation, a, b);
}
```

```
var outputGates = InitialGates.Keys.Where(key => key.StartsWith("z"));

HashSet<string> NeededGates = [..outputGates];
Dictionary<string, bool> Values = InitialValues.ToDictionary(p => p.Key, p => p.Value);

HashSet<string> GatesToAdd = [];
HashSet<string> GatesToRemove = [];
while (NeededGates.Overlaps(outputGates)) {
    foreach(string gateName in NeededGates) {
        Gate gate = InitialGates[gateName];
        if (Values.TryGetValue(gate.a, out bool aValue) && 
            Values.TryGetValue(gate.b, out bool bValue)
        ) {
            GatesToRemove.Add(gateName);
            Values[gateName] = gate.op switch {
                Operation.AND => aValue && bValue,
                Operation.OR  => aValue || bValue,
                Operation.XOR => aValue ^  bValue
            };
        } else {
            if (!Values.ContainsKey(gate.a)) {
                GatesToAdd.Add(gate.a);
            }
            if (!Values.ContainsKey(gate.b)) {
                GatesToAdd.Add(gate.b);
            }
        }
    }
    NeededGates.UnionWith(GatesToAdd);
    NeededGates.ExceptWith(GatesToRemove);
}
string binary = String.Join("", outputGates.Order().Reverse().Select(gate => Values[gate] ? '1' : '0'));
display(binary);
Convert.ToInt64(binary, 2)
```

```
Dictionary<string, Gate> Gates = InitialGates.ToDictionary(p => p.Key, p => p.Value);

void swap(string a, string b) => (Gates[a], Gates[b]) = (Gates[b], Gates[a]);

swap("z12", "qdg");
swap("fgn", "dck");
swap("z37", "nvh");
swap("vvf", "z19");
```

```
var xNodes = InitialValues.Keys.Where(v => v.StartsWith("x")).OrderDescending();
var yNodes = InitialValues.Keys.Where(v => v.StartsWith("y")).OrderDescending();
var pairs = xNodes.Zip(yNodes);
Dictionary<string, HashSet<string>> graphGroups = [];

bool GateIsFrom(Gate gate, string x, string y) => (gate.a == x && gate.b == y) || (gate.b == x && gate.a == y);

foreach(var pair in pairs) {
    var (xNode, yNode) = pair;
    string bit = xNode.Substring(1);
    HashSet<string> group = [xNode, yNode];
    string zNode = $"z{bit}";
    group.Add(zNode);
    group.UnionWith(Gates.Where(gate => GateIsFrom(gate.Value, xNode, yNode)).Select(gate => gate.Key));
    Gate z = Gates[zNode];
    group.Add(z.a);
    group.Add(z.b);
    List<Operation> targetOperations = [Operation.XOR, Operation.OR];
    var inGates = group.Where(node => (node != zNode) && Gates.ContainsKey(node) && targetOperations.Contains(Gates[node].op));
    if (inGates.Count() == 2) {
        group.UnionWith(Gates.Where(gate => GateIsFrom(gate.Value, inGates.First(), inGates.ElementAt(1))).Select(gate => gate.Key));
    }
    
    graphGroups[bit] = group;
}

graphGroups["06"]
```
