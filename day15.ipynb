```
#!import coord.ipynb

Regex mapRegex = new(@"^.*\n\n", RegexOptions.Singleline);
Regex programRegex = new(@"[v<>^]+", RegexOptions.Singleline);

string mapString = mapRegex.Match(data).Value.Trim();
string programString = String.Join("", programRegex.Matches(data).Select(m => m.Value));

enum Direction {
    Up = 0,
    Right = 1,
    Down = 2,
    Left = 3
}
Coord[] Step = [(-1, 0), (0, +1), (+1, 0), (0, -1)];

string[] stringMap = mapString.Split("\n");
char[][] initialMap = stringMap.Select(line => line.ToCharArray()).ToArray();

int width = stringMap[0].Length;
int height = stringMap.Length;

List<Direction> program = programString.Select(c => c switch {
    '^' => Direction.Up,
    '>' => Direction.Right,
    'v' => Direction.Down,
    '<' => Direction.Left,
    _ => throw new System.ArgumentException()
}).ToList();
```

```
using State = (char[][] Map, Coord Position);

Coord ProcessStep(State state, Direction direction) {
    var map = state.Map;
    var pos = state.Position;

    char MapAt(Coord coord) => map[coord.R][coord.C];
    void Swap(Coord a, Coord b) => (map[a.R][a.C], map[b.R][b.C]) = (map[b.R][b.C], map[a.R][a.C]);
    void TriSwap(Coord a, Coord b, Coord c) =>
        (map[a.R][a.C], map[b.R][b.C], map[c.R][c.C]) = (map[c.R][c.C], map[a.R][a.C], map[b.R][b.C]);

    // Is the state valid?
    if (MapAt(pos) != '@') {
        throw new System.InvalidOperationException("Bad state");
    }
    Coord vector = Step[(int)direction];
    Coord target = Add(pos, vector);

    // What's in that direction?
    // Possibilities:
        // It's empty
        // There's a box
        // There's a wall
    switch(MapAt(target)) {
        case '.':
            Swap(pos, target);
            return target;
        case '#':
            // Nope!
            return pos;
        case 'O':
            Coord pushed = target;
            while (MapAt(pushed) == 'O') {
                pushed = Add(pushed, vector);
            }
            if (MapAt(pushed) == '.') {
                TriSwap(pos, target, pushed);
                return target;
            } // Implicit else: What if we just don't?
            return pos;
        default:
            throw new System.InvalidOperationException("Bad map");
    }
}

Coord FindBot(char[][] map) {
    for(int r = 0; r < height; r++) {
        for(int c = 0; c < width; c++) {
            if (map[r][c] == '@') {
                return (r, c);
            }
        }
    }
    throw new System.InvalidOperationException("No Robot?!");
}

// Deep clone
char[][] map = initialMap.Select(a => a.ToArray()).ToArray();

State state = (map, FindBot(map));
// display(state);
foreach (var step in program) {
    state.Position = ProcessStep(state, step);
}
// display(state);
```

```
int GetValue(State state) {
    var map = state.Map;
    int value = 0;
    for(int r = 0; r < height; r++) {
        for(int c = 0; c < width; c++) {
            if (map[r][c] == 'O') {
                value += (r * 100) + c;
            }
        }
    }
    return value;
}

GetValue(state)
```

```
Coord ProcessStep(State state, Direction direction) {
    var map = state.Map;
    var pos = state.Position;

    char MapAt(Coord coord) => map[coord.R][coord.C];

    void Swap(Coord a, Coord b) => (map[a.R][a.C], map[b.R][b.C]) = (map[b.R][b.C], map[a.R][a.C]);
    void Slide(Coord start, Coord end) {
        if (start.R != end.R) throw new InvalidOperationException();
        int r = start.R;
        int sign = Math.Sign(start.C - end.C);
        char endChar = MapAt(end);
        for (int c = end.C; c != start.C; c += sign) {
            map[r][c] = map[r][c + sign];
        }
        map[r][start.C] = endChar;
    }

    // Is the state valid?
    if (MapAt(pos) != '@') {
        throw new System.InvalidOperationException("Bad state");
    }
    Coord vector = Step[(int)direction];
    Coord target = Add(pos, vector);

    bool CanMoveInDirection(Coord boxHalf, Direction direction) {
        // This isn't what this is for.
        if (direction == Direction.Left || direction == Direction.Right) return true;

        Coord otherHalf = MapAt(boxHalf) switch {
            '[' => (boxHalf.R, boxHalf.C+1),
            ']' => (boxHalf.R, boxHalf.C-1),
            _ => throw new NotSupportedException(),
        };

        Coord vector = Step[(int)direction];
        Coord myTarget = Add(boxHalf, vector);
        Coord otherTarget = Add(otherHalf, vector);
        
        return (MapAt(myTarget), MapAt(otherTarget)) switch {
            ('#',  _ ) => false,
            ( _ , '#') => false,
            ('.', '.') => true,
            ('.',  _ ) => CanMoveInDirection(otherTarget, direction),
            ( _ , '.') => CanMoveInDirection(myTarget, direction),
            ( _ ,  _ ) => CanMoveInDirection(myTarget, direction) && CanMoveInDirection(otherTarget, direction)
        };
    }

    void MoveInDirection(Coord boxHalf, Direction direction) {
        // This isn't what this is for.
        if (direction == Direction.Left || direction == Direction.Right) return;

        (Coord leftHalf, Coord rightHalf) = (MapAt(boxHalf) switch {
            '[' => (boxHalf, (boxHalf.R, boxHalf.C+1)),
            ']' => ((boxHalf.R, boxHalf.C-1), boxHalf),
            _ => throw new NotSupportedException(),
        });

        Coord vector = Step[(int)direction];
        Coord lTarget = Add(leftHalf, vector);
        Coord rTarget = Add(rightHalf, vector);
        
        switch(MapAt(lTarget), MapAt(rTarget)) {
            case ('[', ']'):
                MoveInDirection(lTarget, direction);
                break;
            case (']', '['):
                MoveInDirection(lTarget, direction);
                MoveInDirection(rTarget, direction);
                break;
            case (']',  _ ):
                MoveInDirection(lTarget, direction);
                break;
            case ( _ , '['):
                MoveInDirection(rTarget, direction);
                break;
            case ('.', '.'):
                break;
        };
        Swap(leftHalf, lTarget);
        Swap(rightHalf, rTarget);
    }

    // What's in that direction?
    // Possibilities:
        // It's empty
        // There's a box
        // There's a wall
    switch(MapAt(target)) {
        case '.':
            Swap(pos, target);
            return target;
        case '#':
            // Nope!
            return pos;
        case '[':
        case ']':
            if (direction == Direction.Left || direction == Direction.Right) {
                Coord pushed = target;
                while ("[]".Contains(MapAt(pushed))) {
                    pushed = Add(pushed, vector);
                }
                if (MapAt(pushed) == '.') {
                    // TODO We need a slightly fancier push.
                    Slide(pos, pushed);
                    return target;
                } // Implicit else: What if we just don't?
                return pos;
            } else {
                // Oh no. We are vertical. This is actually tree-like.
                if (CanMoveInDirection(target, direction)) {
                    MoveInDirection(target, direction);
                    Swap(pos, target);
                    return target;
                }
                // There is a wall anywhere in the target path. We cannot push.
                return pos;
            }
        default:
            throw new System.InvalidOperationException("Bad map");
    }
}
```
