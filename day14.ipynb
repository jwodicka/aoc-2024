# Part 1

```
using Coord = (int X, int Y);
public static Coord Add(Coord a, Coord b) => (a.X + b.X, a.Y + b.Y);
public static Coord Mult(Coord c, int n) => (c.X * n, c.Y * n);
public static Coord Mod(Coord a, Coord b) => (a.X % b.X, a.Y % b.Y);
```

```
using Robot = (Coord P, Coord V);
```

```
Regex robotRegex = new(@"^p=(\d+),(\d+) v=(-?\d+),(-?\d+)$", RegexOptions.Multiline);
List<Robot> robots = [];
foreach (Match match in robotRegex.Matches(data)) {
    Coord p = (int.Parse(match.Groups[1].Value), int.Parse(match.Groups[2].Value));
    Coord v = (int.Parse(match.Groups[3].Value), int.Parse(match.Groups[4].Value));
    robots.Add((p, v));
}
```

```
int width = 101;
int height = 103;
Coord size = (width, height);

robots = robots.Select(r => 
    (r.P,
    (r.V.X > 0 ? r.V.X : width + r.V.X,
    r.V.Y > 0 ? r.V.Y : height + r.V.Y
    ))
    ).ToList();

var finalStates = robots.Select(r => Mod(Add(r.P, Mult(r.V, 100)), size));

finalStates
```

```
const int halfWidth = 50; //width / 2;
const int halfHeight = 51; // height / 2;

var quads = finalStates.Select<Coord, int?>(s => s switch {
    { X: halfWidth } => null,
    { Y: halfHeight } => null,
    { X: < halfWidth, Y: < halfHeight } => 1,
    { X: < halfWidth, Y: > halfHeight } => 2,
    { X: > halfWidth, Y: < halfHeight } => 3,
    { X: > halfWidth, Y: > halfHeight } => 4,
});

var seenQuads = quads.Where(q => q is not null).Distinct();
var quadCounts = seenQuads.Select(q => quads.Count(v => v == q));
quadCounts.Aggregate(1, (acc, next) => acc * next)
```
