### Imports

```
using System.Collections.Generic;
using System.Text.RegularExpressions;
using System.Linq;
```

# Part 1

```
using Coord = (int R, int C);
enum Facing {
    Up = 0,
    Right = 1,
    Down = 2,
    Left = 3
}
Coord[] Step = [(-1, 0), (0, +1), (+1, 0), (0, -1)];

string[] stringMap = data.Split("\n");
char[][] map = stringMap.Select(line => line.ToCharArray()).ToArray();

int width = stringMap[0].Length;
int height = stringMap.Length;

Coord FindGuard() {
    for (int r = 0; r < height; r++) {
        for (int c = 0; c < width; c++) {
            if (map[r][c] == '^') {
                return (r, c);
            }
        }
    }
    throw new ArgumentException("No guard?!");
}
```

```
Coord guardPos = FindGuard();
Facing guardFacing = Facing.Up;
bool[,] visited = new bool[height,width];

void Mark(Coord c) {
    visited[c.R, c.C] = true;
}

bool IsOnMap(Coord c) {
    if (c.R < 0 || c.C < 0) return false;
    if (c.R >= height || c.C >= width) return false;
    return true;
}

public static Coord Add(Coord a, Coord b) => (a.R + b.R, a.C + b.C);

Facing Turn(Facing from) {
    if (from < Facing.Left) return from + 1;
    return Facing.Up;
}

while(IsOnMap(guardPos)) {
    Mark(guardPos);
    Coord nextPos = Add(guardPos, Step[(int)guardFacing]);
    if (IsOnMap(nextPos)) {
        if (map[nextPos.R][nextPos.C] == '#') {
            guardFacing = Turn(guardFacing);
            continue;
        }
    }
    guardPos = nextPos;
}

int totalVisited = 0;
for (int r = 0; r < height; r++) {
    for (int c = 0; c < width; c++) {
        totalVisited += visited[r, c] ? 1 : 0;
    }
}

Print(totalVisited);
```

# Part 2

```
// The obstruction has to be in a visited space, or it won't matter.

// To detect a loop, we need to know if the guard has ever visited this space *with this facing*.

bool IsLoop(Coord newObstacle, Coord startingPos, Facing startingFacing = Facing.Up) {
    Coord guardPos = startingPos;
    Facing guardFacing = startingFacing;

    HashSet<(Coord Pos, Facing Facing)> seen = [];

    while(IsOnMap(guardPos)) {
        (Coord Pos, Facing Facing) state = (guardPos, guardFacing);
        if (seen.Contains(state)) {
            return true; // This is a loop!
        }
        seen.Add(state);
        
        Coord nextPos = Add(guardPos, Step[(int)guardFacing]);
        if (IsOnMap(nextPos)) {
            if (nextPos == newObstacle || map[nextPos.R][nextPos.C] == '#') {
                guardFacing = Turn(guardFacing);
                continue;
            }
        }
        guardPos = nextPos;
    }
    return false;
}

Coord startingPos = FindGuard();
List<Coord> loops = [];

for (int r = 0; r < height; r++) {
    for (int c = 0; c < width; c++) {
        Coord pos = (r, c);
        if (visited[r, c] && pos != startingPos) {
            if (IsLoop(pos, startingPos)) {
                loops.Add(pos);
            };
        }
    }
}

Print(loops.DebugPrint());
Print(loops.Count);
```
