# Part 1

```
using Equation = (System.Int128 Result, System.Int128[] Numbers);

List<Equation> equations = [];

foreach (string line in data.Split("\n")) {
    string[] parsed = line.Split(": ");
    var result = System.Int128.Parse(parsed[0]);
    var numbers = parsed[1].Split(" ").Select(s => System.Int128.Parse(s)).ToArray();
    equations.Add((result, numbers));
}
```

```
bool CanBeValid(Equation e) {
    Int128 result = e.Result;
    Int128[] numbers = e.Numbers;

    List<Int128> results = [];
    foreach (int number in numbers) {
        if (results.Count == 0) {
            results = [number];
        } else {
            results = [
                ..results.Select(r => r + number),
                ..results.Select(r => r * number)
            ];
        }
    }

    return results.Contains(result);
}

public static Int128 Sum(this IEnumerable<Int128> self) {
    Int128 acc = 0;
    foreach (var i in self) {
        acc += i;
    }
    return acc;
}

var goodEquations = equations.Where(CanBeValid);
Print(goodEquations.Select(e => e.Result).Sum());
```

```
bool CanBeValidToo(Equation e) {
    Int128 result = e.Result;
    Int128[] numbers = e.Numbers;

    List<Int128> results = [];
    foreach (int number in numbers) {
        if (results.Count == 0) {
            results = [number];
        } else {
            results = [
                ..results.Select(r => r + number),
                ..results.Select(r => r * number),
                ..results.Select(r => Int128.Parse($"{r}{number}"))
            ];
        }
    }

    return results.Contains(result);
}

var gooderEquations = equations.Where(CanBeValidToo);
Print(gooderEquations.Select(e => e.Result).Sum());
```
