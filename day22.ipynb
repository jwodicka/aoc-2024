```
IEnumerable<BigInteger> SecretSequence(int startValue) {
    BigInteger value = startValue;
    while(true) {
        yield return value;
        value = NextValue(value);
    }
}

IEnumerable<int> PriceSequence(int startValue) => SecretSequence(startValue).Select(n => (int)(n % 10));

IEnumerable<int> DeltaSequence(int startValue) {
    foreach (var (pre, post) in PriceSequence(startValue).Pairwise()) {
        yield return post - pre;
    }
}
```

```
/**
    * Get a new enumerable that iterates pairwise through the members
    * of this enumerable.
    *
    * \param loopLast True if the enumeration represents a closed loop
    *                 and the final value should be paired to the first.
    *
    * Pairwise iterator extension from
    * https://stackoverflow.com/questions/577590/pair-wise-iteration-in-c-or-sliding-window-enumerator
    * Answer by James Holwell, 2013
*/
public static IEnumerable<T[]> NWise<T>(this IEnumerable<T> enumerable, int n) {
    Queue<T> seenValues = [];

    foreach (T value in enumerable) {
        seenValues.Enqueue(value);
        if (seenValues.Count > n) {
            seenValues.Dequeue();
        }
        if (seenValues.Count == n) {
            yield return seenValues.ToArray();
        }
    }
}
```

```
IEnumerable<(int[] Sequence, int Value)> SequenceValueMap(int startPrice) =>
    DeltaSequence(startPrice).NWise(4).Zip(PriceSequence(startPrice).Skip(4));
```
